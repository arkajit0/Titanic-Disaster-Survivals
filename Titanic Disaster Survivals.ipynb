## Introduction
#### In this project we will build a classification model where according to different features we can tell if a person has survived or not in titanic disaster.
#### The dataset used has 12 columns that we will divide into 11 features and 1 target column
***The Columns are Given as:***

- PassengerId: This is the onboard passenger's Id Number
- Survived: Person survived or not(0: has not survived, 1: has survived)
- Pclass: Passenger Class  
- Name: Passenger Name
- Sex: Passenger Gender
- Age: Passenger Age
- SibSp: Number of Siblings/Spouses Aboard
- Parch: Number of Parents/Children Aboard
- Ticket: Ticket Number
- Fare: Passenger Fare (British pound)
- Cabin: Cabin Code
- Embarked: Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

## Feature Engineering

```python
#function to fill null values
def impute_nan(df, feature):
    df[feature+'_mean'] = df[feature].fillna(value=np.mean(df[feature].dropna())) #filling with mean
    df[feature+'_median'] = df[feature].fillna(value=np.median(df[feature].dropna())) #filling with median
    df[feature+'_mode'] = df[feature].fillna(value=stats.mode(df[feature].dropna())[0][0]) #filling with mode
    #randomly filling up missing datas
    df[feature+'_random'] = df[feature]
    random_variable = df[feature].dropna().sample(df[feature].isnull().sum(), random_state=0)
    random_variable.index = df[df[feature].isnull()].index
    df.loc[df[feature].isnull(), feature+'_random'] = random_variable
```

```python
#filling missing values in column age
impute_nan(data, 'Age')
```

```python
#visualizing age column after filling the missing spots
plt.figure(figsize=(7,5))
data.Age.plot(kind='kde', color='blue')
data.Age_mean.plot(kind='kde', color='violet')
data.Age_median.plot(kind='kde', color='green')
data.Age_mode.plot(kind='kde', color='yellow')
data.Age_random.plot(kind='kde', color='red')
plt.legend()
```

Here we can clearly see that filling data randomly didn't effect the graph at all

```python
#dropping the rest of Age columns and and other columns
data_tuned = data.drop(['Age', 'Age_mean', 'Age_median', 'Age_mode', 'PassengerId', 'Name', 'Cabin', 'Embarked', 'Ticket'], axis=1)
```

```python
#checking for outliers
fig, ax = plt.subplots(figsize=(20,25))
sns.boxplot(data=data_tuned, width=0.5, ax=ax, fliersize=3)
```

```python
#checking for skewness of data
for column in data_tuned.columns:
    if data_tuned[column].dtype != object:
        print("{} = {}".format(column, data_tuned[column].skew()))
```

we can see that some data are not skewed so we will do some transformation to make the data skewed

```python
#plot to check skewness quality
def plot_data(data, feature):
    plt.figure(figsize=(15,20), facecolor='white')
    plt.subplot(2,2,1)
    data[feature].hist()
    plt.subplot(2,2,2)
    stats.probplot(data[feature], dist='norm', plot=pylab)
    plt.show()
```

```python
print(data_tuned['Fare'].skew())
plot_data(data_tuned, 'Fare')
```

```python
data_tuned['Fare_log'] = np.log1p(data_tuned['Fare'])
print(data_tuned['Fare_log'].skew())
plot_data(data_tuned, 'Fare_log')
```

```python
data_tuned['Fare_reciprocal'] = 1/(data_tuned['Fare']-1)
print(data_tuned['Fare_reciprocal'].skew())
plot_data(data_tuned, 'Fare_reciprocal')
```

```python
data_tuned['Fare_sqrt'] = np.sqrt(data_tuned['Fare'])
print(data_tuned['Fare_sqrt'].skew())
plot_data(data_tuned, 'Fare_sqrt')
```

```python
data_tuned['Fare_exp'] = (data_tuned['Fare']**(2/9))
print(data_tuned['Fare_exp'].skew())
plot_data(data_tuned, 'Fare_exp')
```

```python
data_tuned['Fare_boxcox'], parameter = stats.boxcox(data_tuned['Fare']+0.5)
print('parameter = ', parameter)
print(data_tuned['Fare_boxcox'].skew())
plot_data(data_tuned, 'Fare_boxcox')
```

```python
#dropping rest of Fare column and storing it
data_transformed = data_tuned.drop(['Fare_log', 'Fare', 'Fare_reciprocal', 'Fare_sqrt', 'Fare_exp'], axis=1)
```

```python
#standarizing the data
X = data_transformed.drop(['Survived'], axis=1)
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
#checking for any multicollinearity
variable = X_scaled

vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variable, i) for i in range(variable.shape[1])]
vif['feature'] = X.columns
vif
```

```python
#joining sibling and parch to make total passengers
data_transformed['total_passengers'] = data_transformed['SibSp']+data_transformed['Parch']+1
```

```python
#checking the first five datas
data_transformed.head()
```

```python
#individual counts 
data_transformed['total_passengers'].value_counts()
```

```python
#drop the columns that were used to join together
data_transformed.drop(['Parch', 'SibSp'], axis=1, inplace=True)
```

```python
#using onehot encoding in total passenger column
data_revised_passenger = pd.get_dummies(data_transformed['total_passengers'], prefix='total_passenger', drop_first=True) 
```

```python
#storing the final data
data_final = pd.concat([data_transformed.drop(['total_passengers'], axis=1), data_revised_passenger], axis=1)
```

```python
#looking at the data
data_final.head()
```

## Model Building

```python
#taking x and y as variables
X = data_final.drop(['Survived'], axis=1)
y = data_final['Survived']
```

```python
#standarizing the data
X_scaled = sc.fit_transform(X)
```

```python
#splitting training and testing data and creating an instance of Logistic Regression
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=119)
lr = LogisticRegression()
lr.fit(X_train, y_train)
print('train score = ', lr.score(X_train, y_train))
print('test score = ', lr.score(X_test, y_test))
y_pred_lr = lr.predict(X_test)
conf_mtx = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(conf_mtx, fmt='', annot=True, cmap='rainbow_r')
TP = conf_mtx[0][0]
FP = conf_mtx[0][1]
FN = conf_mtx[1][0]
TN = conf_mtx[1][1]
accuracy = (TP+TN)/(TP+TN+FP+FN)
recall = TP/(TP+FN)
precision = TP/(TP+FP)
f1_score = 2*precision*recall/(precision+recall)
print("accuracy = ", accuracy)
print("recall = ", recall)
print("precision = ",precision)
print("F1_score = ", f1_score)
```

our model is good as it is giving a precision of almost 86%

```python
#saving the model
import pickle
pickle.dump(lr, open('final_model_v1.pkl', 'wb'))
```

```python
#classification report
print(classification_report(y_test, y_pred_lr, target_names=['Not Survived', 'Survived']))
```

```python
#calculating area under curve
auc = roc_auc_score(y_test, y_pred_lr)
print('auc = ',auc)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_lr)
plt.plot(fpr, tpr, color='orange', label='ROC')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--',label='ROC curve (area = %0.2f)' % auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
```

```python
#creating an instance of DecisionTree and fitting the model
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print('train score = ', clf.score(X_train, y_train))
print('test score = ', clf.score(X_test, y_test))
y_pred_clf = clf.predict(X_test)
conf_mtx = confusion_matrix(y_test, y_pred_clf)
sns.heatmap(conf_mtx, fmt='', annot=True, cmap='rainbow')
TP = conf_mtx[0][0]
FP = conf_mtx[0][1]
FN = conf_mtx[1][0]
TN = conf_mtx[1][1]
accuracy = (TP+TN)/(TP+TN+FP+FN)
recall = TP/(TP+FN)
precision = TP/(TP+FP)
f1_score = 2*precision*recall/(precision+recall)
print("accuracy = ", accuracy)
print("recall = ", recall)
print("precision = ",precision)
print("F1_score = ", f1_score)
```

Clearly we can see that our model is overfitting

### Using PCA to reduce columns to improve the model

```python
#Reducing dimension with PCA
from sklearn.decomposition import PCA
pca = PCA()
principalcomponents = pca.fit_transform(X_scaled)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of components")
plt.ylabel("variance (%)")
plt.title('Explained Variance')
```

```python
#compressing the dataset columns
pca = PCA(n_components=9)
new_data = pca.fit_transform(X_scaled)
principal_x = pd.DataFrame(new_data, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9'])
```

```python
#checking the first five rows
principal_x.head()
```

```python
#splitting training and testing data and creating an instance of DecisionTree
X_train, X_test, y_train, y_test = train_test_split(principal_x, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print('train score = ', clf.score(X_train, y_train))
print('test score = ', clf.score(X_test, y_test))
y_pred_clf = clf.predict(X_test)
conf_mtx = confusion_matrix(y_test, y_pred_clf)
sns.heatmap(conf_mtx, fmt='', annot=True, cmap='gist_rainbow')
TP = conf_mtx[0][0]
FP = conf_mtx[0][1]
FN = conf_mtx[1][0]
TN = conf_mtx[1][1]
accuracy = (TP+TN)/(TP+TN+FP+FN)
recall = TP/(TP+FN)
precision = TP/(TP+FP)
f1_score = 2*precision*recall/(precision+recall)
print("accuracy = ", accuracy)
print("recall = ", recall)
print("precision = ",precision)
print("F1_score = ", f1_score)
```

still our model is overfitting

### Hyperparameter Tuning

```python
#defining the parameters
grid_params = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': range(1, 30, 1),
    'min_samples_split': range(2,10,1),
    'min_samples_leaf': range(1,10,1)
}
```

```python
#initializing gridsearchcv
clf_grid = GridSearchCV(estimator=clf, param_grid=grid_params, cv=5, n_jobs=-1, verbose=3)
clf_grid.fit(X_train, y_train)
```

```python
#best score and parameter for the model
print(clf_grid.best_score_)
print(clf_grid.best_params_)
```

```python
#fitting the model with the above parameters
clf = DecisionTreeClassifier(**clf_grid.best_params_)
clf.fit(X_train, y_train)
print('train score = ', clf.score(X_train, y_train))
print('test score = ', clf.score(X_test, y_test))
y_pred_clf = clf.predict(X_test)
conf_mtx = confusion_matrix(y_test, y_pred_clf)
sns.heatmap(conf_mtx, fmt='', annot=True, cmap='rainbow')
TP = conf_mtx[0][0]
FP = conf_mtx[0][1]
FN = conf_mtx[1][0]
TN = conf_mtx[1][1]
accuracy = (TP+TN)/(TP+TN+FP+FN)
recall = TP/(TP+FN)
precision = TP/(TP+FP)
f1_score = 2*precision*recall/(precision+recall)
print("accuracy = ", accuracy)
print("recall = ", recall)
print("precision = ",precision)
print("F1_score = ", f1_score)
```

we have got a stable model with a precision of 85% and the model is not overfitting
